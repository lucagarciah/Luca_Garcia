## Realizando consultas ao banco de dados 

## Simulando a retirada de dados de um banco de dados SQL e jogando em um dataframe Pandas. 

```python
import sqlite3
import pandas as pd

# caminho do arquivo do banco de dados
db_path = "movies.db"

#conectar ao banco de dados 

conn = sqlite3.connect(db_path)

#consulta SQL

query = ''' SELECT * FROM movies''' 

movies = pd.read_sql_query(query, conn) 
```
